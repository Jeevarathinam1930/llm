# Insurance Claims Dataset Generator
This notebook generates a sample insurance claim dataset for use in LLM fine-tuning or validation tasks.

```

import json

# Sample insurance claim validation dataset
data = [
    {
        "input": "Patient underwent knee surgery and was hospitalized for 5 days. The policy covers orthopedic surgeries and hospitalization up to 7 days.",
        "output": "Valid claim. Covered surgery and within hospitalization limits."
    },
    {
        "input": "Patient received dental implants. The policy does not cover dental procedures.",
        "output": "Invalid claim. Dental procedures are excluded."
    },
    {
        "input": "Patient submitted two identical claims for the same MRI scan under the same policy number.",
        "output": "Invalid claim. Duplicate detected."
    },
    {
        "input": "Claim for emergency treatment following a road accident. Policy covers emergency treatment under accidental injuries.",
        "output": "Valid claim. Emergency treatment covered under policy."
    },
    {
        "input": "Claim submitted for cosmetic surgery. Policy terms exclude cosmetic and elective procedures.",
        "output": "Invalid claim. Procedure not covered under the policy."
    },
    {
        "input": "Hospital stay of 3 days due to dengue fever. Policy covers hospitalization for vector-borne diseases.",
        "output": "Valid claim. Meets policy terms."
    }
]

# Save as JSON file
file_path = "/content/insurance_claims.json"
with open(file_path, "w") as f:
    json.dump(data, f, indent=2)

print(f"Dataset saved to: {file_path}")
```
